```python
#toss winners
ipl['toss_winner'].value_counts()
```

```python
#extracting the records of team who won when the bat first
bf=ipl[ipl['toss_winner']!=0]
bf.head()
```

```python
#extracting the records of team for toss decision
bf=ipl[ipl['toss_decision']!='field']
bf.head()
```

```python
#creating a plot of win by runs
plt.figure(figsize=(8,5))
plt.hist(bf['win_by_runs'])
plt.title('Distribution of runs')
plt.xlabel('runs')
plt.show()
```

```python
#creating a plot of tp 3 teams winners
plt.figure(figsize=(6,5))
plt.bar(list(ipl['winner'].value_counts()[0:3].keys()),list(ipl['winner'].value_counts()[0:3]))
plt.title('Winners')
plt.xlabel('runs')
plt.show()
```
